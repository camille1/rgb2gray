# RGB2Gray--A basic grayscale image converter

> A simple API for converting an RGB to grayscale image.

```python
#|hide
from nbdev.showdoc import *

```

```python
#|export
from PIL import Image
import numpy as np
import cv2

import matplotlib.pyplot as plt
from io import BytesIO
import requests
```

```python
#|export
def get_gs(img: Image) -> np.array:
    img = img.convert("RGB")
    img_array = np.asarray(img)
    return cv2.cvtColor(np.asarray(img), cv2.COLOR_RGB2GRAY)
```

Here is an example of conversion...

```python
red = Image.new('RGB', (10, 10), color = (255,0,0))
get_gs(red)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,10))
axes[0].imshow(red)
axes[1].imshow(get_gs(red), cmap='gray');
```
