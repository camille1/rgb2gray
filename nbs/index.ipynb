## How to use

An example with `rgb2gray`

```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
result = rgb2gray(Image.new('RGB', (10, 10), color = (255,255,0)))
result
```

```python
plt.imshow(result, cmap='gray')
plt.show();
```
